```python
#Linear Regression
from sklearn.linear_model import LinearRegression
import seaborn as sns
import statsmodels.api as sm

df = sns.load_dataset('tips')
print(df)
X = df['total_bill'].values.reshape(-1, 1)
print(X)
y = df['tip'].values
print(y)

model = LinearRegression()
model.fit(X, y)

residuals = model.predict(X) - y
sm.qqplot(residuals, line='s')
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 6, 8])

model = LinearRegression()
model.fit(X.reshape(-1, 1), y)

plt.scatter(X,y, color='red', label='Data')
plt.plot(X,model.predict(X.reshape(-1, 1)), color='blue', label='Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
def mse(y_true,y_pred):
  return np.mean((y_true-y_pred)**2)
y_true=np.array([1,2,3,4,5])
y_pred=np.array([1.2,1.8,2.5,3.2,4.1])
mse_value=mse(y_true,y_pred)
print(mse_value)
```

```python
import numpy as np

def gradient_descent(X, y, learning_rate=0.01, num_iterations=1000):
    m, c = 0, 0
    n = len(y)

    for _ in range(num_iterations):

        y_pred = m * X + c


        m_gradient = (-2/n) * np.sum(X * (y - y_pred))
        c_gradient = (-2/n) * np.sum(y - y_pred)
        m -= learning_rate * m_gradient
        c -= learning_rate * c_gradient

    return m, c

X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

m_final, c_final = gradient_descent(X, y)
print(f"Final Slope (m): {m_final:.4f}")
print(f"Final Intercept (c): {c_final:.4f}")
```

```python
# R-squared() measures the proprtion of variance explained by model
# Adjusted R-squared accounts for the number of predictors in the model
# Bias is error due to simpler models results in underfitting
# Variance is due to overly complex models results in overfitting
```
