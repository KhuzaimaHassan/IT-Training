```python
#Create dataframe from a dictionary

import pandas as pd
import numpy as np

employees = {
    'Emp_ID': [101,102,103,104,105],
    'Name': ["hamza",'Talha','jawad','Honda','yamaha'],
    'Age': [24,56,30, np.nan, 29],
    'Salary': [50000, 60000,750000, 80000, np.nan],
    'Department': ['HR','IT','Finance', 'IT','HR'],
    'Location' :['NewYork','islamabad', 'Karachi', 'Lahore', 'Hyderabad']
}
# print(employees)
df = pd.DataFrame(employees)
print(df)
```

```python
#Handing Missing data
df['Age'].fillna(df['Age'].mean(), inplace=True)
# df['Salary'].fillna(df['Salary'].mean(), inplace=True)
df.dropna(subset=['Salary'], inplace=True)

print(df)
```

```python
#filter Employees
it_employees = df[df['Department'] == 'IT']
print('IT Employees:]\n ',it_employees)
```

```python
finance_employees = df[df['Department'] == 'Finance']
print('Finance Employees:]\n ',finance_employees)
```

```python
#Sorting employees by Salary
df['Expereince'] = np.where(df['Salary'] > 70000, 'Senior', 'Junior')
# print(df)
sorted_df = df.sort_values(by='Salary', ascending=False)
print(sorted_df)
```

```python

new_hires = {
    'Emp_ID': [106,107],
    'Name': ["Frank",'John'],
    'Age': [44,30],
    'Salary': [55000, 65000],
    'Department': ['HR','Finance'],
    'Location' :['France','Germany']

}
df_newhire = pd.DataFrame(new_hires)
print(df_newhire)
```

```python
df_final = pd.concat([sorted_df, df_newhire], ignore_index=True)
print(df_final)
```

```python
df_final['Experience'] = np.where(df_final['Salary'] > 70000, 'Senior', 'Junior')
print(df_final)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=[1,2,3,4,5]
y=[10,20,25,30,40]
plt.plot(x,y,marker='o',linestyle='-',color='b')
plt.title('simple line plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
fig,ax=plt.subplots(1,2,figsize=(10,5))
ax[0].plot([1,2,3,4,5],[10,20,25,30,40],marker='o',linestyle='--',color='r')
ax[0].set_title('simple line plot')
ax[0].set_xlabel('x axis')
ax[0].set_ylabel('y axis')
#second subplot
ax[1].bar(['A','B','C','D'],[5,6,7,8], color = 'b')
ax[1].set_xlabel('X-axis')
ax[1].set_ylabel('Y-axis')
ax[1].set_title('Second Subplot')
plt.tight_layout()
plt.show()

plt.savefig('plot1.png',dpi=300,bbox_inches='tight')
```

```python
x=np.linspace(0,10,10)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1,label='sin(x)',linestyle='-',marker='o',color='r')
plt.plot(x,y2,label='cos(x)',linestyle='--',marker='x',color='b')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin(x) and cos(x)')
plt.grid(True)
plt.legend()
plt.savefig('plot2.png',dpi=300,bbox_inches='tight')
```

```python
x=np.random.rand(100)
y=np.random.rand(100)
categories=np.random.randint(0,3,100)
colors=['r','g','b']
plt.scatter(x,y,c=[colors[i] for i in categories])
plt.xlabel('x')
plt.ylabel('y')
plt.title('scatter plot')
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
markers=['o','s','^']
for i ,marker in enumerate(markers):
    plt.scatter(x,y,marker=marker,label=f'category {i+1}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('scatter plot')
plt.legend()
```

```python
plt.scatter(x,y,alpha=0.5,color='purple')
plt.xlabel('x')
plt.ylabel('y')
plt.title('scatter plot with transparency')
```

```python
size=np.random.rand(50)*500
plt.scatter(x,y,s=size,alpha=0.9,color='orange')
plt.xlabel('x')
plt.ylabel('y')
plt.title('scatter plot with size')
```

```python
z=np.random.rand(50)*100
plt.scatter(x,y,s=size,alpha=0.9,c=z,cmap='viridis')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('scatter plot with color')
```

```python

x = np.linspace(0, 10, 50)
y = 3*x + np.random.randn(50) * 5
plt.scatter(x, y, color='blue', label="Data")

m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color='red', label="Trend Line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.title("Scatter Plot with Regression Trend Line")
plt.show()
```

```python
sns.regplot(x=x,y=y,scatter_kws={'color':'blue'},line_kws={'color':'red'})
plt.xlabel('x')
plt.ylabel('y')
plt.title('scatter plot with seaborn regression line')
```
